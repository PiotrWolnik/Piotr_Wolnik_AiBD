```python
# In order to process data we'll use pandas library
import pandas as pd
```

```python
# Original data
original_data = pd.read_csv("../Original Data/tb.csv")
original_data
```

```python
# We're going to drop missing values
no_missing_values = original_data.dropna(subset=['iso2', 'year'])

# Let's keep those rows that have at least 5 not NaN values
no_missing_values = no_missing_values.dropna(axis=0, thresh=5)

# Let's drop some columns that are clearly full of NaN values
no_missing_values.drop(['new_sp_m04', 'new_sp_f04', 'new_sp_m514', 'new_sp_f514', 'new_sp_fu', 'new_sp_mu'], inplace=True, axis=1)

# Let's save filtered data to Analysis Data directory
no_missing_values.to_csv("../Analysis Data/tb_without_missing_values.csv", index=False)

no_missing_values
```

```python
# First we're going to replace all NaN values with 0
df = no_missing_values.fillna(0)

# Then we make iso2 and year identifier variables and other columns we treat as measured variables
df = pd.melt(df, id_vars=['iso2', 'year'], value_vars=list(df.columns)[3:],
             var_name='column', value_name='cases')

df['cases'] = df['cases'].astype(int)
df['sex'] = df['column'].str[7]
df['age'] = df['column'].str[8:].map({
    '014': '0-14',
    '1524': '15-24',
    '2534': '25-34',
    '3544': '35-44',
    '4554': '45-54',
    '5564': '55-64',
    '65': '65+'
})
df = df[['iso2', 'year', 'sex', 'age', 'cases']]
df = df.sort_values(by=['iso2','year', 'age','sex'])

df.reset_index(drop=True)

# We save tidied data to tidied_tb.csv file in Analysis Data subfolder
df.to_csv("../Analysis Data/tidied_tb.csv")

# We print just first 20 rows
df.head(20)
```
